# Bundle pricing

```python
import pandas as pd

names = ["Alyona", "Betty", "Celine", "Delilah"]
bundle_price = [130, 140, 170]
good_1_price = [30, 40, 70, 80]
good_2_price = [60, 100, 140]

alyona_reservation = [30, 140, 170]
betty_reservation = [40, 100, 140]
celine = [80, 60, 140]
Delilah = [70, 60, 130]
reservation = [[30, 140, 170],
              [40, 100, 140],
              [80, 60, 140],
              [70, 60, 130]]

count = 0
profit = 0

price_and_profit = {'Price scheme':[],'Price 1':[], 'Price 2':[], 'Bundle price':[], 'Profit':[]}

for price_b in bundle_price:
    for price_1 in good_1_price:
        for price_2 in good_2_price:
            count += 1
            price_and_profit['Price scheme'].append(count)
            price_and_profit['Price 1'].append(price_1)
            price_and_profit['Price 2'].append(price_2)
            price_and_profit['Bundle price'].append(price_b)
            
            print(f'\nPrice combination {count}:\n')
            print(f'Bundle price is {price_b}, Price of good 1 is {price_1}, and good 2 is {price_2}\n')
            for person_i in range(len(names)):
                bundle_utility = reservation[person_i][2] - price_b
                single_utility = max(0, (reservation[person_i][0] - price_1)) + max(0, (reservation[person_i][1] - price_2))
                if (bundle_utility >= single_utility) and (bundle_utility >= 0):
                    print(f"{names[person_i]} chooses the bundle deal\n")
                    profit += price_b
                    continue
                elif (single_utility > bundle_utility) and (single_utility >= 0):
                    if reservation[person_i][0] >= price_1:
                        print(f"{names[person_i]} opts for good 1\n")
                        profit += price_1
                    if reservation[person_i][1] >= price_2:
                        print(f"{names[person_i]} opts for good 2\n")
                        profit += price_2
            print(f'Profit is {profit}\n\n')
            price_and_profit['Profit'].append(profit)
            profit = 0
# print(price_and_profit)
profit_table = pd.DataFrame(price_and_profit).set_index("Price scheme")
print(profit_table.head(36))
max_profit = profit_table.iloc[profit_table['Profit'].idxmax()]
print(max_profit)


                
            
```
